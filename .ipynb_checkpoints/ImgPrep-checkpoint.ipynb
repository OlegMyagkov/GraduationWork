```python
os.chdir("C:/Users/Administrator/Documents/BigData_learning/Graduation_project/GitHub")
wdir = os.getcwd()
print("сейчас тут: " + wdir)
#os.chdir("./GitHub")
os.chdir("./data_set_2")
d_set = os.getcwd()
print("теперь тут: " + d_set)
for something in os.listdir():     
       if os.path.isdir(something):
            i=1
            #print("рисунок тут:" + os.getcwd()+"/"+something) 
            pic = os.getcwd()+"/"+something
            #print(pic)
            image = imageio.imread(os.getcwd()+"/"+something + "/thumbnail_128x128.png")
            image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            ret, image = cv2.threshold(image, 20, 255, 0)
            #print("BW_image_inv:")
            #ia.imshow(threshold_image_inv)
            #cv2.imwrite(d_set + '/thumbnail_128x128_'+ i + '.png', threshold_image_inv)
            for i in range(2):
                #print(i)
                seq = iaa.Sequential([
                iaa.Affine(rotate=(-3, 3)),
                iaa.Crop(percent=(0, 0.001))
                ], random_order=True)
                images_aug = seq(image=image)
                ret, images_aug = cv2.threshold(images_aug, 20, 255, cv2.THRESH_BINARY_INV)
                ia.imshow(images_aug)
                print(d_set + '/thumbnail_128x128_' + str(i) +'.png')                
                cv2.imwrite(os.getcwd()+"/"+something +'/thumbnail_128x128_' + str(i) +'.png', images_aug)
            os.remove(os.getcwd()+"/"+something +'/thumbnail_128x128.png')
#os.chdir("../")
#print(os.getcwd())








```
